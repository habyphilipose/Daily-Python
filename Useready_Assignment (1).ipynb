```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re

# Download NLTK data (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load train.csv data
df_train = pd.read_csv('train.csv')

# Preprocessing function
def preprocess_text(text):
    text_lower = text.lower()
    text_cleaned = re.sub(r'[^\w\s]', '', text_lower)
    words = word_tokenize(text_cleaned)
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    stemmed_words = [stemmer.stem(word) for word in filtered_words]
    lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
    return ' '.join(lemmatized_words)  # Returning lemmatized words, you can choose stemmed_words if you prefer stemming

# Apply preprocessing to text columns
df_train['File Name'] = df_train['File Name'].apply(preprocess_text)
df_train['Party One'] = df_train['Party One'].apply(preprocess_text)
df_train['Party Two'] = df_train['Party Two'].apply(preprocess_text)

# Display preprocessed data
print(df_train.head())  # Displaying the first few rows of the DataFrame after preprocessing
```

```python
df_train.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine text data if needed (e.g., combining 'File Name', 'Party One', and 'Party Two')
df_train['Combined Text'] = df_train['File Name'] + ' ' + df_train['Party One'] + ' ' + df_train['Party Two']

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Adjust max_features as needed

# Fit and transform the combined text data using TF-IDF vectorization
X_tfidf = tfidf_vectorizer.fit_transform(df_train['Combined Text'])
```

```python
df_train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Assuming 'Agreement Value' is the target variable
y_agreement_value = df_train['Aggrement Value']

# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_tfidf, y_agreement_value, test_size=0.2, random_state=42)

# Initialize and train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate model performance on validation set (if needed)
val_predictions = model.predict(X_val)
# Calculate evaluation metrics like RMSE, R^2, etc.
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Assuming 'model' is your trained Linear Regression model
val_predictions = model.predict(X_val)

# Calculate Root Mean Squared Error (RMSE)
rmse = mean_squared_error(y_val, val_predictions, squared=False)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_val, val_predictions)

# Calculate R-squared score
r_squared = r2_score(y_val, val_predictions)

print("RMSE:", rmse)
print("MAE:", mae)
print("R-squared Score:", r_squared)
```

###RMSE (Root Mean Squared Error): 901.71

The RMSE measures the average deviation of predicted values from actual values. In your case, an RMSE of 901.71 means that, on average, your model's predictions are approximately 901.71 units away from the actual target values. Lower RMSE values indicate better model performance, as they reflect smaller errors between predicted and actual values.

###MAE (Mean Absolute Error): 649.13

The MAE measures the average absolute difference between predicted values and actual values. An MAE of 649.13 means that, on average, your model's predictions are approximately 649.13 units away from the actual target values. Similar to RMSE, lower MAE values indicate better model performance.

###R-squared Score: -0.445

The R-squared score (coefficient of determination) indicates the proportion of variance in the target variable that is explained by the model. A negative R-squared score (-0.445 in your case) suggests that your model performs worse than a model that simply predicts the mean of the target variable. Ideally, the R-squared score should be closer to 1 for a good model, indicating a higher proportion of variance explained.
Based on these metrics:

The RMSE and MAE values indicate that your model's predictions have a relatively high average error compared to the actual values. Lower RMSE and MAE values would be desirable.
The negative R-squared score suggests that your model's predictions are not significantly better than simply predicting the mean of the target variable. This indicates that the model may not capture the underlying patterns in the data well or may be overfitting.

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split

# Assuming X_train, y_train, X_val, y_val are your training and validation data

# Example 1: Random Forest Regression
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
rf_val_predictions = rf_model.predict(X_val)

# Calculate RMSE, MAE, R-squared for Random Forest model
rf_rmse = mean_squared_error(y_val, rf_val_predictions, squared=False)
rf_mae = mean_absolute_error(y_val, rf_val_predictions)
rf_r_squared = r2_score(y_val, rf_val_predictions)

print("Random Forest - RMSE:", rf_rmse)
print("Random Forest - MAE:", rf_mae)
print("Random Forest - R-squared Score:", rf_r_squared)

# Example 2: Support Vector Regression (SVR)
svr_model = SVR()
svr_model.fit(X_train, y_train)
svr_val_predictions = svr_model.predict(X_val)

# Calculate RMSE, MAE, R-squared for SVR model
svr_rmse = mean_squared_error(y_val, svr_val_predictions, squared=False)
svr_mae = mean_absolute_error(y_val, svr_val_predictions)
svr_r_squared = r2_score(y_val, svr_val_predictions)

print("SVR - RMSE:", svr_rmse)
print("SVR - MAE:", svr_mae)
print("SVR - R-squared Score:", svr_r_squared)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Define the parameter grid for Random Forest hyperparameters
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of trees
    'max_depth': [None, 10, 20],      # Maximum depth of trees
    'min_samples_split': [2, 5, 10],  # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4]     # Minimum samples required at each leaf node
}

# Create a Random Forest regressor
rf_model = RandomForestRegressor(random_state=42)

# Perform grid search with cross-validation
grid_search = GridSearchCV(rf_model, param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Train a Random Forest model with the best hyperparameters
best_rf_model = RandomForestRegressor(**best_params, random_state=42)
best_rf_model.fit(X_train, y_train)

# Evaluate the model on the validation set
val_predictions = best_rf_model.predict(X_val)
rmse = mean_squared_error(y_val, val_predictions, squared=False)
mae = mean_absolute_error(y_val, val_predictions)
r_squared = r2_score(y_val, val_predictions)

print("Best Random Forest Model - RMSE:", rmse)
print("Best Random Forest Model - MAE:", mae)
print("Best Random Forest Model - R-squared Score:", r_squared)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor

# Define the parameter grid for Random Forest hyperparameters
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Create a Random Forest regressor
rf_model = RandomForestRegressor(random_state=42)

# Perform randomized search with cross-validation
random_search = RandomizedSearchCV(rf_model, param_distributions=param_grid, n_iter=100, scoring='neg_mean_squared_error', cv=5, random_state=42)
random_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = random_search.best_params_

# Train a Random Forest model with the best hyperparameters
best_rf_model = RandomForestRegressor(**best_params, random_state=42)
best_rf_model.fit(X_train, y_train)

# Evaluate the model on the validation set
val_predictions = best_rf_model.predict(X_val)
rmse = mean_squared_error(y_val, val_predictions, squared=False)
mae = mean_absolute_error(y_val, val_predictions)
r_squared = r2_score(y_val, val_predictions)

print("Best Random Forest Model - RMSE:", rmse)
print("Best Random Forest Model - MAE:", mae)
print("Best Random Forest Model - R-squared Score:", r_squared)
```
